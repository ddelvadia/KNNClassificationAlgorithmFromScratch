```python
# find K value
def findK(train_df):
    length=len(train_df)
    k=math.ceil(math.sqrt(length))
    
    return k
```

```python
# find the distance between single train data to the test data
def findDistance(test_d, train_d):
    total=0.0
    for k in range(len(train_d)):
        total += (train_d[k]-test_d[k])**2
        d=math.sqrt(total)
    return d
```

```python
# sort distance values from lowest to highest using bubbleSort
def bubbleSort(distAndLabel):  
    
    for i in range(len(distAndLabel)):
        for j in range(len(distAndLabel)-1):
            if ( distAndLabel.iloc[j,0] > distAndLabel.iloc[j+1,0] ):
                distAndLabel.iloc[j,:], distAndLabel.iloc[j+1,:] = distAndLabel.iloc[j+1,:], distAndLabel.iloc[j,:]
            
    return distAndLabel
```

```python
# let's predict
def prediction(distAndLabel, k):
    
    sort_by_dist=bubbleSort(distAndLabel)  # this line sorts the distance from lowest to highest
      
    kSorted=sort_by_dist.iloc[0:k,:] #this line picks top K lowest distances    
    
    mostK = kSorted.iloc[:,1].value_counts().reset_index().values.tolist()
    pred=mostK[0][0]

    
    return pred    
```
